```python
def visualize_gradient_descent(x, y, m, b, learning_rate, epochs):
    colors = cm.viridis(np.linspace(0, 1, epochs))
    plt.scatter(x, y)
    plt.plot(x, lr.predict(x), color='black', label=f'OLS = {lr.intercept_}')
    for i in range(epochs):
        loss_slope = -2*np.sum(y-m*x.ravel()-b)
        loss_slope

        step_size = loss_slope*learning_rate
        
        b = b - step_size
        
        y_pred = (m*x+b).reshape(10)
        plt.plot(x, y_pred, color=colors[i], label=f'b={b}')
    
    plt.legend()
    plt.show()
```

```python
m = 75.40
b = 5000
lRate = 0.01
epochs = 100
visualize_gradient_descent(x, y,m=m, b=b, learning_rate=lRate, epochs=epochs )
```
