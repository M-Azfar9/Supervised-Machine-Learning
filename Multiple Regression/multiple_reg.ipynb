# Multiple Regression
More than 1 inputs

```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=10000, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
data = pd.DataFrame({'f1':x[:,0],'f2':x[:,1],'target':y})
```

```python
data.shape
```

```python
fig = px.scatter_3d(data_frame=data, x='f1', y='f2', z='target')
fig.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
r2_score(y_test, y_pred)
```

```python
x1 = np.linspace(-5, 5, 10)
y1 = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(y1, x1)
final = np.vstack((xGrid.ravel().reshape(1,100),yGrid.ravel().reshape(1,100))).T

z_final = lr.predict(final).reshape(10,10)

z = z_final
```

```python
import plotly.graph_objects as go
fig = px.scatter_3d(data, x='f1', y='f2', z='target')

fig.add_trace(go.Surface(x = x1, y = y1, z =z ))

fig.show()
```
