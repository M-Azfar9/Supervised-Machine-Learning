1. It helps in finding the feature selection
2. It helps in finding the model interpretability 
    - It helps you understand how your model makes decisions. This is especially important in domains like:

        1. Healthcare (which symptoms(columns) matter most?)
        2. Finance (what factors influence loan approval?)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'F:\Coding\Gen Ai\Machine Learning\Supervised\Softmax (Multinomial Logistic Reg)\mnist.csv')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
rf.feature_importances_
```

```python
sns.heatmap(rf.feature_importances_.reshape(28,28))
```
